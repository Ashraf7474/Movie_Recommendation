```python
import pandas as pd
```

```python
df=pd.read_csv("Movie_Cleaned_Data.csv")
```

```python
df
```

```python

# Sample input values
input_year = int(input("Enter the year: "))  # Replace with user input
input_genre = input("Enter the genre: ")  # Replace with user input
input_certificate = input("Enter the certificate: ")  # Replace with user input
input_stars = input("Enter the stars: ")  # Replace with user input

# Filter movies based on user input
filtered_movies = df[
    (df['Year of release'] == input_year) &
    (df['Genre'].str.contains(input_genre, case=False)) &  # Use 'str.contains' for partial match
    (df['Certificate'] == input_certificate) |
    (df['Stars'].str.contains(input_stars, case=False))  # Use 'str.contains' for partial match
]

# Sort the filtered movies by a relevant criterion (e.g., Rating or MetaScore)
sorted_movies = filtered_movies.sort_values(by=['Rating'], ascending=False)

# Select the top 15 movies from the sorted list
top_15_movies = sorted_movies.head(15)

# Extract the desired columns (image_link, movie_name, year, genre, stars, description)
recommended_movies = top_15_movies[['Image Link', 'Name of movie', 'Year of release', 'Genre', 'Stars', 'Description']]

# Print or return the recommended movies
recommended_movies


```
